```python
from data_utils import count


hours_online_work_counts: dict[str, int] = count(selected_data["hours_online_work"])
print(f"hours_online_work: {hours_online_work_counts}")

ls_effective_counts: dict[str, int] = count(selected_data["ls_effective"])
print(f"lesson_time: {ls_effective_counts}")

programming_effective_counts: dict[str, int] = count(selected_data["programming_effective"])
print(f"Programming_effectiveness: {programming_effective_counts}")

```

The count function uses the data from the selected_data dictionary(from the select function), to make a dictionary that counts the number of times a spcific data value is used, For example, the number of students who answered that it takes them "0 to 2" hours to complete online work was 23.

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(12, 5), sharey=True)


axes[0].set_title("Hours of Online Work")
axes[0].bar(hours_online_work_counts.keys(), hours_online_work_counts.values())
axes[0].tick_params(axis='x', labelrotation = 45)

axes[1].set_title("Lesson Effectiveness")
axes[1].bar(ls_effective_counts.keys(), ls_effective_counts.values())

axes[2].set_title("Programming Effectiveness")
axes[2].bar(programming_effective_counts.keys(), programming_effective_counts.values())
```
